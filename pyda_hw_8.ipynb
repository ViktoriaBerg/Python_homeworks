```python
"""Задание 1
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
Под временем жизни понимается разница 
между максимальным и минимальным значением столбца timestamp для данного значения userId"""
```

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings_grouped = ratings.groupby('userId').count().reset_index()
ratings_grouped.head()
```

```python
ratings_filtered = ratings_grouped[ratings_grouped['rating']>100]
ratings_filtered.head(2)
```

```python
film_fans_user_ids = ratings_filtered['userId'].tolist()
film_fans_user_ids[:5]
```

```python
ratings_1 = ratings[ ratings['userId'].isin(film_fans_user_ids)]
ratings_1.head()
```

```python
min_max_ratings = ratings_1.groupby('userId').agg(['min', 'max'])['timestamp'].reset_index()
min_max_ratings.head(15)
```

```python
lifetime = ((min_max_ratings['max'] - min_max_ratings['min'])/len(lifetime)).round()
lifetime.head()
```

```python
"""Дана статистика услуг перевозок клиентов компании по типам.
Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента"""
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
table_1=client_base.merge(rzd, how='left', on='client_id').merge(air, how='left').merge(auto, how='left')
```

```python
final_table_1 = table_1[['client_id','rzd_revenue','auto_revenue', 'air_revenue']]
final_table_1.fillna(0)
```

```python
final_table_2 = final_table_1.merge(client_base, how='left', on='client_id')
final_table_2.fillna(0)
```
